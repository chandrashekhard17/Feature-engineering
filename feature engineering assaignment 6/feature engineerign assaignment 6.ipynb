### **Q1: Pearson Correlation Coefficient**

**Pearson Correlation Coefficient (r)** is calculated using the formula:

\[ r = \frac{\text{Cov}(X, Y)}{\sigma_X \sigma_Y} \]

where:
- \(\text{Cov}(X, Y)\) is the covariance of \(X\) and \(Y\)
- \(\sigma_X\) is the standard deviation of \(X\)
- \(\sigma_Y\) is the standard deviation of \(Y\)

**Steps:**
1. **Calculate the mean of \(X\) (study time) and \(Y\) (exam scores).**
2. **Calculate the covariance between \(X\) and \(Y\).**
3. **Calculate the standard deviations of \(X\) and \(Y\).**
4. **Plug these values into the formula.**

**Interpretation:**
- \( r \) ranges from -1 to 1.
- \( r = 1 \) indicates a perfect positive linear relationship.
- \( r = -1 \) indicates a perfect negative linear relationship.
- \( r = 0 \) indicates no linear relationship.

### **Q2: Spearman's Rank Correlation**

**Spearman's Rank Correlation (ρ)** is calculated using the formula:

\[ \rho = 1 - \frac{6 \sum d_i^2}{n(n^2 - 1)} \]

where:
- \(d_i\) is the difference between the ranks of the \(i\)-th pair of values
- \(n\) is the number of observations

**Steps:**
1. **Rank the values for both variables (sleep and job satisfaction).**
2. **Calculate the differences in ranks (\(d_i\)).**
3. **Square the differences (\(d_i^2\)).**
4. **Sum the squared differences.**
5. **Apply the formula.**

**Interpretation:**
- Similar to Pearson's correlation, ρ ranges from -1 to 1.
- Positive ρ indicates a positive monotonic relationship.
- Negative ρ indicates a negative monotonic relationship.
- ρ close to 0 suggests a weak monotonic relationship.
